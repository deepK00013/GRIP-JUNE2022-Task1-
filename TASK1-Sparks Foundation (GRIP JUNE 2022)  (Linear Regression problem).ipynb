### TASK 1: Prediction using Supervised ML (To predict the percentage of  a student on the basis of his number of study hours)

<font color=maroon>**Importing the required libraries**</font>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<font color=maroon>**Reading data from the given link**</font>

```python
url="http://bit.ly/w-data"
data=pd.read_csv(url)
print("Data imported successfully")

#looking for top 5 values in the dataset
data.head(5)
```

<font color=maroon>**Data visualization**</font>

```python
#plotting the scores w.r.t the number of hours studied by the student
sns.scatterplot(x=data['Hours'],y=data['Scores'],marker='*',s=200,color='green')
plt.title(' Scores v/s Study Hours',size=20)
plt.xlabel('Number of Hours Studied',size=12)
plt.ylabel('Score Percentage',size=12)
plt.grid(True)
plt.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=1)
plt.minorticks_on()
plt.show()
```

**The scatter plot shows that there is a positive linear relation b/w the number of hours studied and score percentage.Further a regression line is drawn to show the correlation**

```python
#plotting the regression line to show correlation
sns.regplot(x=data['Hours'],y=data['Scores'],marker='*',scatter_kws={"color":"green"},
                                                    line_kws={"color":"red"},ci=None)
plt.title(' Scores v/s Study Hours',size=20)
plt.xlabel('Number of Hours Studied',size=12)
plt.ylabel('Score Percentage',size=12)
plt.grid(True)
plt.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=1)
plt.minorticks_on()
plt.show()
print(data.corr())
```

<font color=maroon>**Data Processing**</font>

```python
# Dividing the data into attributes and labels
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

<font color=maroon>**Model Training**</font>

```python
#splitting the data into training and testing datasets using train_test_split method
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=0)
```

<font color=maroon>**Training the Algorithm**</font>

```python
#Now training the Algorithm
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
print("The Algorithm is trained now")
```

```python
# Plotting the regression line for the test data
line=lr.coef_*x+lr.intercept_
plt.scatter(x,y,color="blue")
plt.plot(x,line,color="yellow")
plt.grid(True)
plt.grid(which='minor', color='#EEEEEE', linestyle=':', linewidth=1)
plt.minorticks_on()
plt.show()
```

<font color=maroon>**Model Predictions**</font>

```python
# predicting the scores
y_pred=lr.predict(x_test)
print("Prediction is completed")
```

```python
# Comparison of actual and predicted values
comparison=pd.DataFrame({"Actual values":y_test,"Predicted values":y_pred})
comparison
```

<font color=maroon>**Predicting Scores for given value**</font>

```python
#Predicting the scores if the student studies for 9.25 hours per day
hours=9.25
my_pred=lr.predict([[hours]])
print("No. of study hours={}".format(hours))
print("Predicted Score={}".format(my_pred[0]))
```

**The predicted score indicates that if a student studies for 9.25 hours daily then his expected  percentage score will be 93.7%**

<font color=maroon>**Evaluating the Model Performance**</font>

```python
#Finding the Mean Absolute Error to determine the accuracy of the model
from sklearn import metrics
mae=metrics.mean_absolute_error(y_test,y_pred)
accuracy_train=lr.score(x_train,y_train)
accuracy_test=lr.score(x_test,y_test)
print("Mean Absolute Error:{:.5f}".format(mae))
print("Linear Regression model accuracy on training dataset: {: .5f}".format(accuracy_train))
print("Linear Regression model accuracy on test dataset: {: .5f}".format(accuracy_test),
      "\n\n Evaluation completed....")
```
